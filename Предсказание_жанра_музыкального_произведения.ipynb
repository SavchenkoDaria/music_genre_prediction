# Предсказание музыкального жанра

В нашем распоряжении данные популярного музыкального стримингового сервиса "МиФаСоль", в котором собраны некоторые характеристики музыкальных произведений и их жанры. Наша задача — правильно классифицировать новые музыкальные треки, чтобы улучшить работу рекомендательной системы, то есть, разработать модель, позволяющую классифицировать музыкальные произведения по жанрам.

***Описание данных***

Данные находятся в трёх файлах:

- **train.csv** - информация о музыкальных треках, которые будут использоваться в качестве обучающих данных;
- **test.csv** - информация о музыкальных треках, которые будут использоваться в качестве тестовых данных.

Описание полей данных
- **instance_id** - уникальный идентификатор трека;
- **track_name** - название трека;
- **acousticness** - акустичность;
- **danceability** - танцевальность;
- **duration_ms** - продолжительность в милисекундах;
- **energy** - энергичность;
- **instrumentalness** - инструментальность;
- **key** - базовый ключ (нота) произведения;
- **liveness** - привлекательность;
- **loudness** - громкость;
- **mode** - указывает на модальность (мажорную или минорную) трека;
- **speechiness** - выразительность;
- **tempo** - темп;
- **obtained_date** - дата загрузки в сервис;
- **valence** - привлекательность произведения для пользователей сервиса;
- **music_genre** - музыкальный жанр.

#### Столбец 'key' - базовый ключ (нота) произведения

Для начала выведем все строки данного столбца с пустыми значениями, предварительно посмотрим на данные и выдвенем предположение, каким способом можно заполнить пропуски:

```python
(data_train[data_train['key'].isnull()])
```

Первое, что бросается в глаза - достаточно часто в названии классической музыки написана нота в формате "Serenade **in B flat**, K.361 "Gran partita...". Нотная грамота по-английски выглядит так:
- C - До
- D - Ре
- E - Ми
- F - Фа
- G - Соль
- A - Ля
- B - Си

Значит, в нашем примере 'key' - 'B', а 'flat' - это знак альтерации (бемоль). Забегая вперед, знак альтерации нам, к сожалению, не поможет в заполнении пропущенных значений в столбце 'mode', так как мажорность или минорность определить только по знаку нельзя.

Заполним пропуски, где это возможно:

```python
def my_func(x):
    if re.search(r'[i,I]n\s[C,D,E,F,G,A,B]', x):
        return re.findall(r'[i,I]n\s[C,D,E,F,G,A,B]', x)[0][-1:]
    else:
        np.nan

data_train['key'] = data_train.apply(lambda x: my_func(x['track_name']) if pd.isnull(x['key']) else x['key'], axis = 1)
```

```python
data_test['key'] = data_test.apply(lambda x: my_func(x['track_name']) if pd.isnull(x['key']) else x['key'], axis = 1)
```

Проверим, реализовалась ли наша замена, выведем строку с индексом 87453 - "Serenade in B flat, K.361 "Gran partita..."

```python
(data_train[(data_train['instance_id'] == 87453.0) & (data_train['music_genre'] == 'Classical')])
```

Немного уменьшили количество пропусков, заполнив их корректными значениями в 20 музыкальных произведениях:

```python
data_train['key'].isna().sum()
```

Посмотрим, какую долю занимают пропуски от общего числа значений:

```python
l = data_train['key'].isna().sum()
k = data_train['key'].shape[0]
round((l/k)*100,2)
```

3.51% - небольшой процент, пропуски будут заменены перед подбором гиперпараметров методом `RandomForestClassifier()`.

#### Проверка модели на мультиколлинеарность

Визуализируем признаки нашего датафрейма и определим, какие их них сильно коррелируют друг с другом:

## Обучение моделей

### Определение признаков

В первую очередь зафиксируем численные и категориальные признаки для дальнейшего преобразования:
- **num_col_list** отвечает за численные;
- **cat_col_list** - за категориальные.

```python
num_col_list = ['acousticness',
                'danceability',
                'duration_ms',
                'energy',
                'instrumentalness',
                'liveness',
                'loudness',
                'speechiness',
                'tempo',
                'valence']
cat_col_list = ['key', 'mode']
```

Создадим первый пайплайн, который будет отвечать за преобразование численных признаков - заполнять пустые значения с использованием подхода к-ближайших соседей, а также стандартизировать набор данных.

```python
knn_imputer = KNNImputer(n_neighbors=5, weights='uniform')
std_scaler = StandardScaler()

pipe_num = Pipeline([('imputer', knn_imputer), ('scaler', std_scaler)])
```

Напишем функцию, которая применяет пайплайн для численных столбцов:

```python
def num_features(x_num):
    res_num = pipe_num.fit_transform(x_num[num_col_list])
    res_df_num = pd.DataFrame(res_num, columns=pipe_num['scaler'].get_feature_names_out(x_num[num_col_list].columns))
    return res_df_num
```

### Обработка пропусков в категориальных столбцах

Далее обработаем пропуски в категориальных столбцах 'key' и 'mode' в тренировочной и тестовой выборках. Для этого создадим функцию, которая предскажет значения в пропущенных строках с помощью метода `RandomForestClassifier()`

```python
def for_nan_clf(dataset, col):
    params = {'bootstrap': True,
              'ccp_alpha': 0.0,
              'class_weight': None,
              'criterion': 'entropy',
              'max_depth': 50,
              'max_features': 'sqrt',
              'max_leaf_nodes': None,
              'max_samples': None,
              'min_impurity_decrease': 0.0,
              'min_samples_leaf': 1,
              'min_samples_split': 2,
              'min_weight_fraction_leaf': 0.0,
              'n_estimators': 50,
              'n_jobs': None,
              'oob_score': False,
              'random_state': None,
              'verbose': 0,
              'warm_start': False}
    clf = RandomForestClassifier(**params)
    clf.fit(num_features(dataset.loc[dataset[col].notna()][num_col_list]), dataset.loc[dataset[col].notna()][col])
    return clf
```

Применим функцию и заполним пропущенные значения на тренировочных данных, взяв индексы таких строк:

```python
nan_index = data_train.index[data_train['key'].isna()].tolist()
data_train.loc[data_train['key'].isna(), 'key'] = for_nan_clf(data_train, 'key').predict(num_features(data_train[num_col_list]))[nan_index]

nan_index = data_train.index[data_train['mode'].isna()].tolist()
data_train.loc[data_train['mode'].isna(), 'mode'] = for_nan_clf(data_train, 'mode').predict(num_features(data_train[num_col_list]))[nan_index]
```

Аналогично обработаем пропуски на тестовых данных:

```python
nan_index = data_test.index[data_test['key'].isna()].tolist()
data_test.loc[data_test['key'].isna(), 'key'] = for_nan_clf(data_train, 'key').predict(num_features(data_test[num_col_list]))[nan_index]

nan_index = data_test.index[data_test['mode'].isna()].tolist()
data_test.loc[data_test['mode'].isna(), 'mode'] = for_nan_clf(data_train, 'mode').predict(num_features(data_test[num_col_list]))[nan_index]
```

### Создание пайплайнов

Теперь перейдем к созданию второго пайплайна, который будет использоваться уже для обучения моделей. Нам необходимо выполнить следующие преобразования для численных столбцов:
- заполнить пропуски, если они имеются;
- масштабировать данные;
- создать полиномиальные признаки из набора данных;
- удалить признаки, которые сильно коррелируют между собой.

```python
poly = PolynomialFeatures()
drop_correlation = DropCorrelatedFeatures(threshold=0.9)

pipe_num = Pipeline([('imputer', knn_imputer),
                     ('scaler', std_scaler),
                     ('poly', poly),
                     ('drop', drop_correlation)])
```

В то время как для категориальных признаков мы применим только порядковое кодирование:

```python
o_encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

pipe_cat = Pipeline([('encoder', o_encoder)])
```

Объединим пайплайны воедино:

```python
col_transformer = ColumnTransformer([('num_preproc', pipe_num, num_col_list),
                                    ('cat_preproc', pipe_cat, cat_col_list)])
```

И отделим целевой признак - это столбец 'music_genre', так как мы предсказываем жанр трека. В тестовых данных этого столбца нет, поэтому для удобства разграничения перезапишем все столбцы тестового датафрейма в переменную features_test.

```python
features_train = data_train.drop(columns='music_genre')
target_train = data_train.music_genre

features_test = data_test
```

Следующим шагом будет применения нашего пайплайна к тренировочным данным:

```python
res = col_transformer.fit_transform(features_train)
res_df = pd.DataFrame(res, columns = [x.split('__')[-1] for x in col_transformer.get_feature_names_out()])
```

### Поиск лучших гиперпараметров с Optuna

#### Метод CatBoostClassifier()

С помощью фреймворка **Optuna** автоматизируем подбор гиперпараметров для модели `CatBoostClassifier()`:

```python
def run_cb(trial):
    param = {
        'learning_rate': trial.suggest_float('learning_rate', 1e-3, 0.1, log=True),
        'depth': trial.suggest_int("depth", 3, 12),
        'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1e-3, 1.0, log=True),
        'bootstrap_type': trial.suggest_categorical("bootstrap_type", ["Bayesian"]),
        'random_strength': trial.suggest_float("random_strength", 1e-8, 10.0, log=True),
        'bagging_temperature': trial.suggest_float("bagging_temperature", 0.0, 10.0),
        'od_type': trial.suggest_categorical("od_type", ["IncToDec", "Iter"]),
        'od_wait': trial.suggest_int("od_wait", 10, 50)
    }

    cb_model = CatBoostClassifier(**param,
                                 iterations=4000,
                                 random_seed=12345,
                                 objective='MultiClass',
                                 verbose=False,
                                 eval_metric='MultiClass')

    cb_model.fit(
        res_df,
        target_train,
        early_stopping_rounds=100,
        verbose=1000)

    f1 = cross_val_score(cb_model, res_df, target_train, n_jobs=-1, cv=5, scoring='f1_micro').mean()
    return f1


cb_study = optuna.create_study(direction='maximize')
cb_study.optimize(run_cb, n_trials=30, timeout=30)

display(cb_study.best_trial)
display(cb_study.best_params)
display(cb_study.best_value)
```

Для модели, обученной методом `CatBoostClassifier()`, значение F1-меры составило 0.4838547069231245 на тренировочных данных.

#### Метод RandomForestClassifier()

```python
def run_rf(trial):
    param = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 151),
        'max_depth': trial.suggest_int('max_depth', 2, 15),
        'min_samples_split': trial.suggest_int('min_samples_split', 1, 150),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 60),
    }

    rf_model = RandomForestClassifier(**param,
                                      n_jobs=-1,
                                      random_state=12345)

    rf_model.fit(
        res_df,
        target_train)

    f1 = cross_val_score(rf_model, res_df, target_train, cv=5, scoring='f1_micro').mean()
    return f1


rf_study = optuna.create_study(direction='maximize',
                               pruner=optuna.pruners.MedianPruner(n_warmup_steps=5))
rf_study.optimize(run_rf, n_trials=30, timeout=30)

display(rf_study.best_trial)
display(rf_study.best_params)
display(rf_study.best_value)
```

Для модели, обученной методом `RandomForestClassifier()`, значение F1-меры составило 0.43086062911744205 на тренировочных данных.

#### Метод XGBClassifier()

```python
def run_xgb(trial):
    params = {
        'max_depth': trial.suggest_int('max_depth', 1, 9),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.01, 1.0),
        'n_estimators': trial.suggest_int('n_estimators', 50, 500),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
        'gamma': trial.suggest_loguniform('gamma', 1e-8, 1.0),
        'subsample': trial.suggest_loguniform('subsample', 0.01, 1.0),
        'colsample_bytree': trial.suggest_loguniform('colsample_bytree', 0.01, 1.0),
        'reg_alpha': trial.suggest_loguniform('reg_alpha', 1e-8, 1.0),
        'reg_lambda': trial.suggest_loguniform('reg_lambda', 1e-8, 1.0)
    }

    xgb_model = XGBClassifier(**param,
                              n_jobs=-1,
                              objective='multi:softmax',
                              random_state=12345)

    xgb_model.fit(
        res_df,
        target_train)

    f1 = cross_val_score(xgb_model, res_df, target_train, cv=5, scoring='f1_micro').mean()
    return f1


xgb_study = optuna.create_study(direction='maximize')
xgb_study.optimize(run_rf, n_trials=30, timeout=30)

display(xgb_study.best_trial)
display(xgb_study.best_params)
display(xgb_study.best_value)
```

Для модели, обученной методом `XGBClassifier()`, значение F1-меры составило 0.4204573104183039 на тренировочных данных.

#### Метод LGBMClassifier()

```python
def run_lgbm(trial):
    param = {
        'lambda_l1': trial.suggest_float('lambda_l1', 1e-8, 10.0, log=True),
        'lambda_l2': trial.suggest_float('lambda_l2', 1e-8, 10.0, log=True),
        'num_leaves': trial.suggest_int('num_leaves', 2, 256),
        'feature_fraction': trial.suggest_float('feature_fraction', 0.4, 1.0),
        'bagging_fraction': trial.suggest_float('bagging_fraction', 0.4, 1.0),
        'bagging_freq': trial.suggest_int('bagging_freq', 1, 7),
        'min_child_samples': trial.suggest_int('min_child_samples', 5, 100),
    }

    lgbm_model = LGBMClassifier(**param,
                                n_jobs=-1,
                                objective='multiclass',
                                random_state=12345)

    lgbm_model.fit(
        res_df,
        target_train)

    f1 = cross_val_score(lgbm_model, res_df, target_train, cv=5, scoring='f1_micro').mean()
    return f1


lgbm_study = optuna.create_study(direction='maximize')
lgbm_study.optimize(run_rf, n_trials=30, timeout=30)

display(lgbm_study.best_trial)
display(lgbm_study.best_params)
display(lgbm_study.best_value)
```

Для модели, обученной методом `LGBMClassifier()`, значение F1-меры составило 0.4764393526086007 на тренировочных данных.
